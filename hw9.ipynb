```python
x=np.array([250., 500, 1000, 2000, 3000, 4000])
y=np.array([ 50000. , 100000, 200000, 400000, 600000, 800000])
randomNumbers = np.random.normal(0.0, 1.0, 6)
randomNumbers = np.random.normal(0.0, 1.0, 6)
noise=randomNumbers*30000
y=noise+y
li2=[1 for x in range(len(x))]

x2=np.asarray(x)
xT=np.vstack([li2,np.asarray(x)])
x2=np.transpose(xT)
y2=np.asarray(y).reshape(len(y),1)
xTx=np.matmul(xT, x2)

xTy=np.matmul(xT, y2)

A=np.matmul(np.linalg.inv(xTx), xTy)
yEQ=A[0]+ (A[1] *x)

plt.plot(x,yEQ,color = 'b', label='lines')
plt.scatter(x,y, color = 'r')
plt.title('Least Squares Regression Line')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.legend()
plt.show()

```

```python
import numpy as np
li= [(-2, 3), (-1, 5), (0, 1), (1, 4), (2,10)]
x=[]
y=[]

for r in li:
  for c in range(0,len(r)):
    if(c == 0):
      x.append(r[c])
    else:
      y.append(r[c])
a = [[0 for i in range(len(x))] for j in range(len(x))]
temp=[]
degree=[1+x for x in range(-1,len(li)-1)]
for value in range(0,len(x)):
  for c in range(0,len(degree)):
    temp.append((x[value]**degree[c]))
count=0
a=np.asarray(a)
b=np.asarray(y)
for i in range(len(a)):
  for j in range(len(a[i])):
      a[i][j]= temp[count]
      count=count+1
print(a)
aInv=np.linalg.inv(a)
newy = np.matmul(aInv, b)
print(f'Coefficients: {newy}.')
mymodel = 0
points = np.linspace(min(x), max(x), len(x)*3)
for i in range(len(x)):
    mymodel += newy[i]*(points**i)
plt.plot(x1, mymodel, color = 'b',label ='points')
plt.scatter(x1, mymodel, color = 'b')
plt.title("Polynomial Curve Fitting")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.legend()
plt.show()
```
